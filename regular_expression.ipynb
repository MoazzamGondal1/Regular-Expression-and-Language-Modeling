```python
import nltk
import re
from pprint import pprint
from urllib import request
from bs4 import BeautifulSoup
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
sample_text = f"The quick brown fox jumps over the lazy dog. \
Please pay $50.25 for 3 apples and 2 oranges. \
Python is a powerful programming language. \
The cat sat on the mat. \
a dead poet \
The rain in Spain falls mainly on the plain. \
12345 apple banana pear grape watermelon \
3*8+10"
```

# Question 1

```python
regex_pattern = r'[a-zA-Z]+'
print("(a):", regex_pattern)
nltk.re_show(regex_pattern, sample_text)

bregex_pattern = r'[A-Z][a-z]*'
print("\n(b):", bregex_pattern)
nltk.re_show(bregex_pattern, sample_text)

cregex_pattern = r'p[aeiou]{,2}t'
print("\n(c):", cregex_pattern)
nltk.re_show(cregex_pattern, sample_text)

dregex_pattern = r'\d+(\.\d+)?'
print("\n(d):", dregex_pattern)
nltk.re_show(dregex_pattern, sample_text)

eregex_pattern = r'([^aeiou][aeiou][^aeiou])*'
print("\n(e):", eregex_pattern)
nltk.re_show(eregex_pattern, sample_text)

fregex_pattern = r'\w+|[^\w\s]+'
print("\n(f):", fregex_pattern)
nltk.re_show(fregex_pattern, sample_text)
```

a- The regular expression [a-zA-Z]+ matches one or more occurrences of any alphabetical character (both lowercase and uppercase).

b- The regular expression [A-Z][a-z]* matches a capital letter followed by zero or more lowercase letters.

c- This regular expression matches strings that start with the letter 'p', followed by zero to two vowels, and ends with the letter 't'.

d-  This regular expression matches numbers, including decimal numbers.

e- This regular expression matches sequences of three characters where the middle character is a vowel and the surrounding characters are not vowels. 

f- This regular expression matches either one or more word characters (letters, digits, or underscores) or one or more non-word, non-whitespace characters.

# Question 3

```python
def remove_html_tags(url):
    html_content = request.urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(html_content, 'html.parser')
    return soup.get_text()

url = 'https://www.dsu.edu.pk/contact-us/'
text = remove_html_tags(url)
print(text)
```

# Question 4

```python
print("Phone Numbers in text :")
regex_phoneNumber = r'\b\d{3,4}\-\d+\b'
print(re.findall(regex_phoneNumber,text))

print("\nEmail Addresses in text :")
regex_e = r'[a-zA-Z0-9.]+[@][a-zA-Z.]+'
print(re.findall(regex_e, text))

print("\nText after Tokenization :")
tokens = nltk.word_tokenize(text)
print(tokens)
```

# Question 5

```python
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()

porter_stemmed_text = [porter_stemmer.stem(word) for word in tokens]
lancaster_stemmed_text = [lancaster_stemmer.stem(word) for word in tokens]

print("Text after Porter Stemming :")
print(porter_stemmed_text)

print("\nText after Lancaster Stemming :")
print(lancaster_stemmed_text)
```

```python
regex1 = r'[zZ][a-z]*a'
first_word = re.search(regex1, message).group()
print("Frequency : ",len(re.findall(regex1, message)))
print("First Word : ",first_word)
```

```python
regex2 = r'[0-9]k[a-z]*[0-9]{1}'
second_word = re.search(regex2, message).group()
print("Frequency : ",len(re.findall(regex2, message)))
print("Extracted Word : ",second_word)
second_word = second_word[3:-3]
print("Actual Word : ",second_word)
```

```python
regex3 = r'c[a-z]*\*[a-z]+a'
third_word = re.search(regex3, message).group()
print("Extracted Word : ", third_word)
third_word = third_word.replace('*','')
print("Actual Word :", third_word)
```

```python
regex4 = r'[pP][a-z]{2}a'
fourth_word = re.search(regex4, message).group()
print("Frequency : ",len(re.findall(regex4, message)))
print("Extracted Word : ",fourth_word)
```
